# Vectorization:

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True,stop_words='english',max_features=25000)
vector = cv.fit_transform(new_md['Specs']).toarray()
```

```python
vector  
```

```python
cv.get_feature_names()
```

# Similarity:

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape[0]
```

```python
similarity.shape
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6]
```

```python
new_md[new_md['Model']=='Infinix']
```

```python
new_md[new_md['Model']=='OPPO']
```

```python
def recommend_Model(mobile):
    mobile_index = new_md[new_md['Model']==mobile].index[0]
    distances = similarity[mobile_index]
    mobile_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
    for i in mobile_list:
        print(new_md.iloc[i[0]]['Model'])
```

```python
recommend_Model('Infinix HOT 20 Play (Luna Blue, 64 GB)')
```
